```python
### Code for drawing lines on one image:

# Open Image file
im = Image.open('/home/luispengler/code/projects/signature-rec/001_01.PNG')

# Get image dimensions
x, y = im.size

# Define width of the line
width = 3
w = y - width

# Define where line starts
b = 0

# Create a Drawing Object
draw = ImageDraw.Draw(im)

# Define the line coordinates and color

rect_coordinates = (x, w, b, y)
rect_color = "green"

# Draw the line
draw.rectangle(rect_coordinates, fill=rect_color)

# See image
#plt.imshow(im)
plt.imshow(im)
```
